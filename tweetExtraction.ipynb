```python
examples = [{"TWEET": tweet1,"Answer":"yes"},
 {"TWEET":tweet2,"Answer":"no"},
  {"TWEET":tweet3,"Answer":"yes"},
   {"TWEET":tweet4,"Answer":"no"},
    {"TWEET":tweet5,"Answer":"yes"},
    ]
```

```python
# Create an example template
example_template = """
TWEET : {TWEET},

AI : {Answer}
"""
# Create an example prompt from the above template

example_prompt = PromptTemplate(
    input_variables = ["TWEET","Answer",],
    template = example_template
)

prefix = """Given a TWEET, tell whether it is related to hinduism or not. Give answer as yes or no."""
suffix = """
{TWEET}
AI: """

#Now create few shot prompt template
few_shot_prompt_template = FewShotPromptTemplate(
    examples = examples,
    example_prompt = example_prompt,
    prefix = prefix,
    suffix = suffix,
    input_variables = ["TWEET"],
    example_separator = "\n\n"
)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(llm_model_name)
tok = AutoTokenizer.from_pretrained(llm_model_name)
```

```python
model.to(device)
# tok.to(device)
```

```python
tok.model_max_length=2048
```

```python
hindutweets2=[]
```

```python
def get_ans(tweet, few_shot_prompt_template):


  # print(few_shot_prompt_template.format(CONTENT=combined_doc, Question = Question))
  # ans1 = flanllm(few_shot_prompt_template.format(CONTENT=combined_doc, Question = Question))
  input_ids = tok(few_shot_prompt_template.format(TWEET=tweet), return_tensors="pt", padding=True)
  out = model.generate(**input_ids, max_new_tokens=100, num_beams=4, do_sample=True)
  ans2 = (tok.batch_decode(out, skip_special_tokens=True))
  if(ans2=="yes"):
    hindutweets2.append(tweet)
```
